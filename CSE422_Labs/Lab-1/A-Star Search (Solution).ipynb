```python
from queue import PriorityQueue


def read(heuristic,graph):
    
    f = open("Input file.txt", 'r')
    
    for line in f:
        i = 2
        adjacent=[]

        nodes=line.split(" ") # List
        
        
        
        nodes[-1]=nodes[-1].strip("\n")
        heuristic[nodes[0]]=int(nodes[1])    # huristic dict has been created in a-star function.
                                             # key te area and value te distance boshano hoyeche.
        
        
        while i <len(nodes) - 1:
            child = (nodes[i], int(nodes[i + 1]))  #child nodes gula ke list e tuple hishebe rekhe dichi
            i=i+2
            adjacent.append(child)    # child and its distance stored in this list.
        graph[nodes[0]]=adjacent      # key te parents and value te child er list with distance.
    
    print(graph)
    print('.')
    print('.')
    print('.')
    print('.')
    
    #print(heuristic)


    f.close()
    
def output(goal,back_node,source,track):
    p=[goal]
    traceback=back_node

    while traceback!=source:
        for k,v in track.items():

            if traceback in v:
                p.append(k[1])

                traceback=k

    print('Path:', p[-1], end="")
    
    
    for i in range(len(p) - 2, -1, -1):
        print(" ->", p[i], end="")
        
    print()
    print("Total distance: "+str(back_node[0])+" km")


def cost(track,graph,temp): 

    recent,mother,p_cost=temp,temp,0

    while mother!=None:
        mother=None
        cost=0
        
        for k,v in track.items():
            
            if recent in v:
                mother=k            # Parent node updated
                break
        
        
        for k,v in graph[recent[1]]:
            
            

            if mother==None:
                break
            if k == mother[1]:
               
                cost=v
                break
                
        p_cost+=cost
        recent=mother

    return p_cost



def a_star():
    
    explored=[]   
    track={} 
    heuristic={}
    graph={}
    
    read(heuristic, graph)
    lst=PriorityQueue() 
    goal = "Bucharest"
      

    s=input("Enter source node: ")
    
    if s not in heuristic.keys():
        print("PATH NOT FOUND")
        
    

    
    back_node=None    # goal node pele sheta store korbo
    
    source = (heuristic[s],s)
    
    # source fn is always equal to its heuristic value
    
    track[source]=[]    # key te parent distance, area and value te list of child distance, area
    
    lst.put(source)     # list e first e source node er tuple entry kora hoyeche
    
    #print(track)
    
    
    


    while not lst.empty():

        node=lst.get()     # here node is a tuple (where distance is at 0 index and area at index 1)
        explored.append(node[1])
        
        print(node)

        # check if popped node is goal node
        
        if node[1] == goal:
            back_node=node
            break
            
        # create a key with the popped node if not in keys
        
        if node not in track.keys():
            track[node] = []

        for k,v in graph[node[1]]:
            
        
            # here k=node and v=cost
            
            # Parent is A and children are B and C
            
            if k in explored:
                continue

            # find cost of the path using track from parent to source
            
            ref= cost(track,graph,node)
            
            # find fn
            
            f=heuristic[k]+v+ref
            
            track[node].append((f,k))      
            lst.put((f,k))

    output(goal,back_node,source,track)
    
    print('.')
    print('.')
    print('.')
    print('.')
    
    print(track)
    


a_star()
```
