```python
import math
import random

pruned_nodes = 0 


def minimax(position, depth, alpha, beta, maximizingPlayer, tree,branch):
    global pruned_nodes  

    temp = branch
    


    if tree[position] is not None:
        return tree[position]

    if maximizingPlayer:
        maxEval = -math.inf
        for i in range(1, temp+1):
            Eval = minimax(branch * position + i, depth - 1, alpha, beta, False, tree, temp)
            maxEval = max(maxEval, Eval)
            tree[position] = maxEval
            alpha = max(alpha, Eval)
            
            if beta <= alpha:
                pruned_nodes += temp-i
                break
        
        return maxEval

    else:
        minEval = math.inf
        for i in range(1,temp+1):
            Eval = minimax(branch * position + i, depth - 1, alpha, beta, True, tree, temp)
            minEval = min(minEval, Eval)
            tree[position] = minEval
            beta = min(beta, Eval)
            
            if beta <= alpha:
                pruned_nodes += temp-i  
                break
                

        return minEval


def modify_ID(minPoints, maxPoints, PointsToWin, d, b):
    global pruned_nodes  
    
    tree = []
    #tree = [None, None, None, None, 19, 22, 9, 2, 26, 16, 16, 27, 16]
    
    g=[]   # leaf node chara baki jei brach gula (excluding the root node)
    ref=1
    temp=1 # leaf node e koto gula node thakbe (basically no. of leaf nodes)
    
    for i in range(1, d):
        ref=ref*b
        g.append(ref)
        
    for j in range(sum(g)+1):
        tree.append(None)
        
    for k in range(d):
        temp=temp*b
        
    for m in range(temp):
        tree.append(random.randint(minPoints, maxPoints))
        

    high = (sum(g)+1)+temp  # total nodes in graph  including leaf and root nodes (denotes last node index of the tree list)
    
    
    
    score = minimax(0, d, -math.inf, +math.inf, True, tree,b)

    print("Terminal States (leaf node values) are", ((str(tree[(sum(g)+1):high])).strip(']')).strip('['))
    print("Left life(HP) of the defender after maximum damage caused by the attacker is", PointsToWin-score)
    print("After Alpha-Beta Pruning Leaf Node Comparisons:", temp-pruned_nodes)

    
input_ID = input("Enter your student ID: ")
points = str(input('Minimum and Maximum value for the range of negative HP:'))
print()
var=points.split(' ')
minPoints=int(var[0])
PointsToWin = int(input_ID[7:5:-1])
maxPoints=int(var[1])
branch = int(input_ID[2])
depth = int(input_ID[0])*2


print('Depth and Branches ratio is '+str(depth)+': '+str(branch))
modify_ID(minPoints, maxPoints, PointsToWin, depth, branch)
```
